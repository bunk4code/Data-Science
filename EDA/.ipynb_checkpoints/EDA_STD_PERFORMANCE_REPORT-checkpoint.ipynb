```python
categorical_data = data[data.dtypes[data.dtypes == 'object'].index]
continuous_data=data[data.dtypes[data.dtypes != 'object'].index]
```

```python

total_marks=data['math score']+data['reading score']+data['writing score']
```

```python
data['total_marks']=total_marks
```

```python
data.head(5)
```

```python
data['average_score'] = data['total_marks']/3
data.head(5)
```

## Visualization
##### exploring more visualization

```python
fig,axis= plt.subplots(1,2,figsize =(15,5))
plt.subplot(121)
sns.histplot(data=data,x='average_score',bins= 30,kde=True,color='g')
plt.subplot(122)
sns.histplot(data=data,x='average_score',bins= 30,kde=True,color='g',hue='gender') 
```

# Insights
##### fig.(2) represents females have more intelligence and readiablity then males
##### both male and female have got highest average marks.
##### males performes bettern in marks of betwen 40 to 65 average score.

```python
fig,axis= plt.subplots(1,3,figsize =(25,6))
plt.subplot(141)
sns.histplot(data=data,x='average_score',bins= 30,kde=True,color='g',hue='lunch')
plt.subplot(142)
sns.histplot(data=data[data['gender']=='female'],x='average_score',bins= 50,kde=True,color='g',hue='lunch') 
plt.subplot(143)
sns.histplot(data=data[data['gender']=='male'],x='average_score',bins= 50,kde=True,color='g',hue='lunch') 
```

# Insights
##### more standard lunch gives more performancy in average score.


```python
fig,axis= plt.subplots(1,3,figsize =(25,6))
plt.subplot(141)
sns.histplot(data=data,x='average_score',bins= 30,kde=True,color='g',hue='parental level of education')
plt.subplot(142)
sns.histplot(data=data[data['gender']=='female'],x='average_score',bins= 50,kde=True,color='g',hue='parental level of education') 
plt.subplot(143)
sns.histplot(data=data[data['gender']=='male'],x='average_score',bins= 50,kde=True,color='g',hue='parental level of education') 
```

## Insights
