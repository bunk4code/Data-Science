# Smote
#### SMOTE (Synthetic Minority Over-Sampling Technique) is a technique use in maching learning to address imbalanced datasets where the minority class has significantly fewere instances than majority class.SMOTE involves generating synthetic instances of mthe minority class by interpolating between existing instances.

```python
from sklearn.datasets import make_classification
```

```python
# X is independent (input)
# Y is dependent (output)

X,y=make_classification(n_samples=1000,n_features = 2,n_redundant=0,n_clusters_per_class=1,weights=[0.90],  random_state=1 )
```

```python
import pandas as pd
df1 = pd.DataFrame(X,columns=['f1','f2'])
df2 = pd.DataFrame(y,columns=['target'])
```

```python
data = pd.concat([df1,df2],axis=1)
```

```python
data.head()
```

```python
data['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(data['f1'],data['f2'],c=data['target'])
```

# OverSampling

```python
from imblearn.over_sampling import SMOTE
```

```python
overSampling = SMOTE()
```

```python
X,y = overSampling.fit_resample(data[['f1','f2']],data['target'])
```

```python

df1 = pd.DataFrame(X,columns=['f1','f2'])
df2 = pd.DataFrame(y,columns=['target'])
final_oversampled_data = pd.concat([df1,df2],axis=1)
```
